```python
df2['season_start'] = df2['Season Start'] 
df2 = df2.drop('Season Start',axis=1)
```

```python
from datetime import datetime
date_test = df2.season_start[1]
date_test = date_test.astype(int)
date_test = date_test.astype(str)
date_test
```

```python
date_dt1 = datetime.strptime(date_test, '%Y')
date_dt1.year
```

```python
def make_year(string):
    date_test = int(string)
    date_test = str(date_test)
    date_year = datetime.strptime(date_test, '%Y')
    return date_year.year
```

```python
make_year(2018.0)
```
