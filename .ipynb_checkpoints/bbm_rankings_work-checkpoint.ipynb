```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
file = 'data/BBM_PlayerRankings.xls'
```

```python
xl = pd.ExcelFile(file)
```

```python
print(xl.sheet_names)
```

```python
df = xl.parse('Sheet 1')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop('Own',axis=1)
```

```python
df.describe()
```

```python
df.fillna(0)
```

```python
df['Inj'] = df['Inj'].map(lambda x: 1 if x == "Injured" else 0)
```

```python
df.iloc[8]
```

```python
df.points_per_g.plot(kind='hist')
```

```python
df['points_per_g'].groupby(df['Pos']).mean().sort_values(ascending= False)
```

```python
df['Round'].groupby(df['Pos']).mean()
```

```python
df['Round'].groupby(df['Team']).mean().sort_values().plot(kind='bar')
```

```python

df['Rank'].groupby(df['Team']).mean().sort_values().plot(kind='barh')
```

```python
plt.matshow(df.corr())
```

```python
corr = df.corr()
```

```python
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)
```

```python
df_subgroup = df[['points_per_g', 'Rank', 'ast_per_g', 'stl_per_g', 'to_per_g', 'fta_per_g', 'fg_pct','fga_per_g']]
```

```python
sns.pairplot(df_subgroup)
```

```python
corr_sub = df_subgroup.corr()
```

```python
sns.heatmap(corr_sub, xticklabels=corr_sub.columns.values, yticklabels=corr_sub.columns.values)
```
